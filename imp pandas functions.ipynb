```python
import numpy as np
import pandas as pd

#creating some dataframes for further use

values_1 = np.random.randint(10, size =10)
values_2 = np.random.randint(10, size =10)
years = np.arange(2010, 2020)
groups = ['A','B','A','C','B','B','A','A','A','C']
year1 = np.arange(2000, 2010)
df = pd.DataFrame({'Group':groups, 'Year':years, 'Value_1': values_1,'Value_2': values_2})
df1 = pd.DataFrame({'Group':groups, 'Year':year1, 'Value_1': values_1,'Value_2': values_2})
```

```python
#using query function to filter data

df.query('Value_1 < Value_2')
```

```python
#using insert function to insert a new column in the dataframe at a desired position instead of at the end

Value_3 = np.random.randn(10)

#inserting value_3 column in position 2 i.e 3 column
df.insert(2,'Value_3', Value_3)
df
```

```python
#using cumsum with group by to find cumulative sum

df['cumsum']=df[['Value_2','Group']].groupby('Group').cumsum()
df
```

```python
#using sample to select a random sample from a series or dataframe

df.sample(n=3)
```

```python
#using 'Where' to replace rows or columns based on a condition

df['Value_3'].where(df['Value_3']>0,0)
```

```python
#Using 'isin' function to filter records based on a selection of values

year = ['2010','2012','2015']
df[df.Year.isin(year)]
```

```python
#using loc and iloc
#loc: selects by labels. Column names are column labels. Row labels are int starting from 0 if no label is specified 
#iloc: selects by position

#select first 3 rows and first 2 columns with iloc:
df.iloc[:3,:2]

#select first 3 rows and first 2 columns with loc. Loc icludes the upper boundary hence :2 will suffice:
df.loc[:2,['Group','Year']]

#select rows 1,3,5 and columsn year and values_1
df.loc[[1,3,5],['Year','Value_1']]
```

```python
#using pct_change to check the change in percentage of each succeeding element in a series

df.Value_1.pct_change()
```

```python
#using 'rank' to assign ranks to values
df['Rank_1'] = df['Value_1'].rank()
df
```

```python
#Using 'Melt' to convert wide dataframes into narrow ones i.e pivoting down columns into rows

df.melt(id_vars=['Group','Year'])
```

```python
#Using nunique to find the number of unique entries over columns and rows

#to check #of unique values in column Year
df.Year.nunique()

#to check # of unique values in each column
df.nunique()
```

```python
#Using infer_objects to check which would be a mor apt datatype for the columns for which datatype is object

df.infer_objects().dtypes
```

```python
#Using memory_usage to check usage of each column

df.memory_usage()
```

```python
#Using describe to determine the statistical information on numerical columns

df.describe()
```

```python
#Using 'Merge' to join two dataframe based on 1 column
#Using the 'how' parameter can help to join in various ways - inner, outer, left, right

#df_merge = pd.merge(df, df1, on='Group')
df_merge = pd.merge(df, df1, how='inner', on = 'Group')
```

```python
#Using select_dtypes to include or exclude certain columns based on the datatype

df.select_dtypes(include='int32')
df.select_dtypes(exclude='int32')
```

```python
#Using 'replace' to update a value in a dataframe

#df.replace('A','A_1')
df.replace({'A':'A_1','B':'B_1','C':'C_1'})
```
